```python
import tensorflow as tf
import os
import cv2
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import confusion_matrix
import seaborn as sns
```

Importamos los datos y los dividimos en Entrenamiento y Validación

```python
mnist = tf.keras.datasets.mnist
(X_train, y_train), (X_test, y_test) = mnist.load_data()
```

Escalamos los datos para que estén en [0-1]

```python
X_train = tf.keras.utils.normalize(X_train, axis= 1)
X_test = tf.keras.utils.normalize(X_test, axis= 1)
```

Creamos el modelo de Red Neuronal, la entrada son imágenes de 28x28, la primer capa flatten hace que la matriz se transforme en un vector.
Esta Red tiene 2 capas ocultas de 16 neuronas cada una y una capa de salida de 10 neurona.

```python
model = tf.keras.models.Sequential()
model.add(tf.keras.layers.Flatten(input_shape= (28, 28)))
#no es un grid es una linea de 784 pixeles
model.add(tf.keras.layers.Dense(16, activation= 'relu'))
# dense son las capas del medio
model.add(tf.keras.layers.Dense(16, activation= 'relu'))
model.add(tf.keras.layers.Dense(10, activation= 'softmax'))
```

Compilamos el modelo utilizando como función de pérdida crossentropy y como métrica accuracy

```python
model.compile(optimizer= 'adam', loss= 'sparse_categorical_crossentropy', metrics= ['accuracy'])
```

Entrenamos el modelo

```python
model.fit(X_train, y_train, epochs= 10)
```

Validamos el modelo con los datos que reservamos para validación

```python
loss,acuracy = model.evaluate(X_test, y_test)

print(loss)
print(acuracy)
```

```python
predictions = model.predict(X_test)
predicted_labels = np.argmax(predictions, axis=1)
conf_matrix = confusion_matrix(y_test, predicted_labels)
# Imprimir la matriz de confusión
print("Matriz de Confusión:")
print(conf_matrix)

# Visualizar la matriz de confusión
plt.figure(figsize=(10, 8))
sns.heatmap(conf_matrix, annot=True, fmt="d", cmap="Blues")
plt.xlabel('Predicted')
plt.ylabel('True')
plt.title('Confusion Matrix')
plt.show()
```
